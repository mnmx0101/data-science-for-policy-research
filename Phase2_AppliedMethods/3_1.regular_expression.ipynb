```python
# =========================================
# Extra Session: Regex for Text Processing
# =========================================

import re

# Example dataset
texts = [
    "Contact us at support@example.com or call (123)-456-7890!",
    "The event is scheduled for 12/25/2024.",
    "Visit https://www.example.org for more info."
]

# ------------------------------
# Part 1: Extract Emails
# ------------------------------
email_pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
for t in texts:
    print("Text:", t)
    print("Emails:", re.findall(email_pattern, t))

# ------------------------------
# Part 2: Extract Dates
# ------------------------------
date_pattern = r"\d{2}/\d{2}/\d{4}"
for t in texts:
    print("Text:", t)
    print("Dates:", re.findall(date_pattern, t))

# ------------------------------
# Part 3: Extract Phone Numbers
# ------------------------------
phone_pattern = r"\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}"
for t in texts:
    print("Text:", t)
    print("Phones:", re.findall(phone_pattern, t))

# ------------------------------
# Part 4: Extract URLs
# ------------------------------
url_pattern = r"https?://[A-Za-z0-9./]+"
for t in texts:
    print("Text:", t)
    print("URLs:", re.findall(url_pattern, t))

# ------------------------------
# Mission Task
# ------------------------------
# 1. Write a regex to extract hashtags (#example) from text.
# 2. Write a regex to find all capitalized words in a sentence.
# 3. Use regex to remove punctuation before passing text to spaCy or Hugging Face models.
```
