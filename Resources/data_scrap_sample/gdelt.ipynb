# Retrieve global events -Daily

```python
for i in range(2010, 2016, 1):
    
    for j in range(1,13, 1):
        
        # Create Empty list to put monthly GDELT Event data
        result_append =[]
        
        for k in range(1, 32, 1):
        
            if j in [1,3,5,7,8,10,12]:

                month = str(i) +' ' + str(j) + ' ' + str(k)
                
                print(month)
            elif j in [2]:
                if k-3 >=1:
                    month = str(i) +' ' + str(j) +' ' +str(k-3)

                    print(month)
            else:
                if k-1 >= 1:
                    month = str(i) +' ' + str(j) +' ' +str(k-1)
                #end = str(i) +' ' + str(j) +' 30'

                    print(month)
                
            file_path = os.getcwd()+r'/gdelt_data/event_'+str(i)+'_' +str(j)+'.csv'            
            
            if not os.path.exists(file_path):
                
                try:
                    results = gd1.Search(month, coverage=True, table='events')

                    print("length of the file :", len(results), results['DATEADDED'].unique())


                    # concat daily event
                    result_append.append(results)

                    print(len(result_append))
                
                except:
                    print('no data')

            #concat monthly event
            result = pd.concat(result_append)
            
            print(len(result))

        result.to_csv(os.getcwd()+r'/gdelt_data/event_'+str(i)+'_' +str(j)+'.csv')
        print('event_'+str(i)+'_' +str(j)+ '- downdload completed')
        
        
```

# Retrieve global events - Monthly

```python
for i in range(2019, 2020, 1):
    
    for j in range(1, 13, 1):
        
        ## set different days by calendar month
        
        # those with 31 days
        if j in [1,3,5,7,8,10,12]:
            
            start = str(i) +' ' + str(j) +' 1'
            end = str(i) +' ' + str(j) +' 30'
            print(start, end)
        
        # those with 28 days
        elif j in [2]:
            
            start = str(i) +' ' + str(j) +' 1'
            end = str(i) +' ' + str(j) +' 28'
            print(start,end)
        
        # others with 30 days
        else:
            start = str(i) +' ' + str(j) +' 1'
            end = str(i) +' ' + str(j) +' 30'
            
            print(start,end)
        
        # Path to the CSV file
        file_path = os.getcwd()+r'/gdelt_data/event_'+str(i)+'_' +str(j)+'.csv'
        
        # Measure time
        start_time = time.time()
        
        # Check if the file exists, create it if it doesn't
        if not os.path.exists(file_path):
            
            results = gd1.Search([start,end], coverage=True, table='events')
            
            print("length of the file :", len(results))
            
            results.to_csv(file_path)
            
            print(f"CSV file for {start, end}Created")
            
        else:
            print("CSV file already exists.")
        
        end_time = time.time()
        
        print('time spent :', end_time-start_time)
        
        # Pause 1 min for every other loop
        time.sleep(60)
```

```python
countrycode = 'NGA'

concat_file = []

filelist = glob.glob(os.getcwd()+r'/gdelt_data/*.csv')

for f in filelist:
    dask_df = dd.read_csv(f, dtype={'Actor1KnownGroupCode': 'object',
                                    'Actor1Type2Code': 'object',
                                    'Actor1Type3Code': 'object',
                                    'Actor2Type3Code': 'object',
                                    'Actor1EthnicCode': 'object',
                                   'Actor1Religion1Code': 'object',
                                   'Actor1Religion2Code': 'object',
                                   'Actor2EthnicCode': 'object',
                                   'Actor2Religion2Code': 'object',
                                   'EventCode':'object',
                                   'EventBaseCode': 'object',
                                   'Actor2Religion1Code': 'object',
                                   'EventRootCode': 'object',
                                   'ActionGeo_Long': 'object',
                                   'Actor1Geo_Long': 'object',
                                   'Actor2Geo_Long': 'object'})
    print(f.split('/')[-1][:-4])
    subdf = dask_df[(dask_df['Actor1CountryCode']==countrycode)|(dask_df['Actor2CountryCode']==countrycode)].compute()
    
    subdf['date'] = pd.to_datetime(subdf['DATEADDED'], format ='%Y%m%d')
    concat_file.append(subdf)
```
